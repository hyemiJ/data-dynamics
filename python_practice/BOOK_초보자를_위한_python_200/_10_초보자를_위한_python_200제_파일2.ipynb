```python
# 접근 권한이 막히는 문제가 있는 경우 해결하기 위해 헤더를 추가한다
from urllib.request import urlopen, Request

imgurl = "https://images.unsplash.com/photo-1503023345310-bd7c1de61c7d"
imgname = "test_image.jpg"

headers = {
    'User-Agent': 'Mozilla/5.0'
}

req = Request(imgurl, headers=headers)

try:
    with urlopen(req) as u:
        with open(imgname, 'wb') as f:
            img = u.read()
            f.write(img)
    print("이미지 다운로드 완료!")
except Exception as e:
    print(f"에러 발생: {e}")
```

```python

url = 'https://speed.hetzner.de/100MB.bin'  # 100MB 테스트 파일
save_name = 'test_file.bin'
BUFSIZE= 256*1024
try:
    with urlopen(url) as u, open(save_name, 'wb') as f:
        while True:
            data = u.read(BUFSIZE)  # 256KB씩 읽기
            if not data:
                break
            f.write(data)
    print("다운로드 완료!")
except Exception as e:
    print(f"에러 발생 : {e}")
```

```python
# ssl 에러를 위한 ssl 설정후 정상 저장되는 것으로 보였으나 파일 크기가 정상적이지 않았음
# 다른 url로 테스트
from urllib.request import urlopen, Request
import ssl
import os

url = 'https://proof.ovh.net/files/100Mb.dat'  # OVH 테스트 서버
save_name = 'test_file.bin'
BUFSIZE = 256 * 1024

context = ssl._create_unverified_context()

req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})

with urlopen(req, context=context) as u, open(save_name, 'wb') as f:
    while True:
        data = u.read(BUFSIZE)
        if not data:
            break
        f.write(data)

print("다운로드 완료!")
print(f"저장된 파일 크기 : {os.path.getsize(save_name)} Byte")
```

```python
import os

file_name = 'test_file.bin'  # 원본 파일 이름
subsize = 1024 * 1024 * 10  # 10MB 단위
suffix = 0

# 원본 파일 크기 확인
print(f"원본 파일 크기 : {os.path.getsize(file_name)} Byte")

with open(file_name, 'rb') as f:
    while True:
        data = f.read(subsize)  # 10MB씩 읽기
        if not data:  # 파일 끝나면 종료
            break
        subname = f"{file_name}_{suffix}"  # 분할 파일명
        with open(subname, 'wb') as f2:
            f2.write(data)
        print(f"{subname} 저장 완료")
        suffix += 1
```

```python
#파일을 zip 압축 파일로 만들기.
from zipfile import *
def compressZip(zipname,filename):
  print(f"{filename}압축 중..")
  with ZipFile(zipname,'w') as ziph:
    ziph.write(filename)
  print(f"{zipname}압축 완료.")
```

```python
file_name = 'test_file.bin'
zip_name = file_name +".zip"
compressZip(zip_name,file_name)
```

```python
#폴더를 zip 압축 파일로 만들기
def compressAll(dir_name):
  zip_name = dir_name + ".zip"
  print(f"{dir_name} -> {zip_name} 압축중..")
  with ZipFile(zip_name,'w') as ziph:
    for dirname, subdirs, files in os.walk(dir_name):
      for file in files:
        ziph.write(os.path.join(dirname,file))
```

```python
folder = "sample_data"
compressAll(folder)
```

```python
def extractZip2(zipname, extract_path="."):
    with ZipFile(zipname, 'r') as ziph:
        # 압축파일 안 파일 목록 가져오기
        zip_file_list = ziph.namelist()
        # 압축 풀기
        ziph.extractall(extract_path)
        print(zipname, "이 성공적으로 추출되었습니다.")

    # 추출된 경로에서 파일 목록 가져오기
    extracted_files = []
    for root, dirs, files in os.walk(extract_path):
        for file in files:
            extracted_files.append(os.path.relpath(os.path.join(root, file), extract_path))

    # zip파일 목록과 추출된 파일 목록 비교
    if set(zip_file_list) <= set(extracted_files):
        print("모든 파일이 성공적으로 추출되었습니다.")
    else:
        print("일부 파일이 추출되지 않았습니다.")

extractZip2("sample_data.zip", extract_path="extracted_folder")
```
